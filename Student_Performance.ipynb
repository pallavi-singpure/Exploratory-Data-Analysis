Outlier detection
1.To Delect outliers
2.Distribution , boxplot ,voilinplot

```python
students_grades_df = students_grades_df.drop(columns=['G1', 'G2','G3'],axis=1)
```

```python
def detect_outliers(columns):
    outlier_indices = []
    for col in columns:
        Q1 = np.percentile(students_grades_df[col], 25)
        Q3 = np.percentile(students_grades_df[col], 75)
        IQR = Q3 - Q1
        mask =(students_grades_df[col] < Q1 - 1.5 * IQR) | (students_grades_df[col] > Q3 + 1.5 * IQR)
        false_indices=np.argwhere(~mask).flatten()
        outlier_indices.extend(false_indices)
    return np.unique(outlier_indices)
```
