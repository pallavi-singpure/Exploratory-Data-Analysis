```python
def size_process(item):
  if str(item)[-1]=='M':
    res = float(str(item)[:-1])
  elif str(item)[-1]=='k':
    res = float(str(item)[:-1])/1024
  else:
    res = np.nan
  return res

size_process('19M')
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
```
