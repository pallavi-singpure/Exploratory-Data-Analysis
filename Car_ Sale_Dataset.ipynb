6)How does the average selling price of cars vary by customer gender and annual income.

```python
df['Price ($)']=pd.to_numeric(df['Price ($)'],errors='coerce')
df['Annual Income']=pd.to_numeric(df['Annual Income'],errors='coerce')
```

```python
# Create income brackets for better visualization
bins = [0, 30000, 60000, 100000, 200000, 500000]
labels = ['Low', 'Lower-Middle', 'Middle', 'Upper-Middle', 'High']
df['Income Bracket'] = pd.cut(df['Annual Income'], bins=bins, labels=labels)
```

```python
# Plot a boxplot to visualize average selling price by gender and income bracket
plt.figure(figsize=(12, 6))
sns.boxplot(x='Gender', y='Price ($)', hue='Income Bracket', data=df, palette='pastel')
plt.xlabel('Gender')
plt.ylabel('Average Selling Price')
plt.title('Average Selling Price by Gender and Income Bracket')
plt.legend(title='Income Bracket')
plt.show()
```

14.What is the correlation between car price, engine size, and annual income of customers, and how do these features interact

```python
df['Engine']=pd.to_numeric(df['Engine'],errors='coerce')
df['Annual Income']=pd.to_numeric(df['Annual Income'],errors='coerce')
df = df.dropna()  # Drop rows with invalid values after conversion
```

```python
df.columns
```

```python
# Calculate and display correlation matrix
corr_matrix = df[['Price ($)', 'Engine', 'Annual Income']].corr()
print("Correlation Matrix:")
print(corr_matrix)
```

```python
# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap: Price, Engine Size, and Annual Income")
plt.show()
```
