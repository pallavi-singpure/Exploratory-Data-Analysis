# EDA- UNIVARITE ANALYSIS

Univariate Analysis is the simplest form of data analysis, where only one variable is analyzed at a time. It helps in understanding the distribution, central tendency, and dispersion of the data.

**Types of Univariate Analysis**
Descriptive Statistics

**Measures of Central Tendency:** Mean, Median, Mode
Measures of Dispersion: Range, Variance, Standard Deviation, Quartiles
Skewness & Kurtosis: Understanding the shape of the distribution
Visualization Techniques

**Histogram**: Shows the frequency distribution of a variable
**Box Plot**: Highlights median, quartiles, and outliers
**Bar Chart**: Used for categorical data representation
**Pie Chart**: Displays proportions

```python
headmap=pd.crosstab(df['Sex'],df['Survived'])
```
